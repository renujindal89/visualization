# Scatter Plots

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
#experiance -> salary
# age-> weight
#cgpa->package
```

```python
import seaborn as sns
```

```python
# size
tips = sns.load_dataset('tips')

tips
```

```python
plt.scatter(tips['total_bill'],tips['tip'])
```

```python
tips.describe()
```

```python
color_map = {'Male': 0, 'Female': 1}
plt.scatter(tips['total_bill'],tips['tip'],c=tips['sex'].map(color_map))
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.show()
```

```python


sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
df = pd.read_csv(r'D:\batsman_season_record.csv')
df
```

```python
## Convert wide  long format
long_df = df.melt(id_vars='batsman', var_name='Year', value_name='Runs')
long_df
```

```python
sns.barplot(data=long_df, x='batsman', y='Runs', hue='Year')

plt.title("Multiple Bar Chart in Seaborn")
plt.show()
```

# pie chart

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects,autopct='%0.0f%%')

plt.show()
```

```python
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects,autopct='%0.1f%%',colors=['blue','green','yellow','pink','brown'])

plt.show()
```
