```python
# size
tips = sns.load_dataset('tips')

tips
```

```python
plt.scatter(tips['total_bill'],tips['tip'])
```

```python
tips.describe()
```

```python
color_map = {'Male': 0, 'Female': 1}
plt.scatter(tips['total_bill'],tips['tip'],c=tips['sex'].map(color_map))
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
color_map = {'Male': 'blue', 'Female': 'red'}

plt.scatter(tips[tips['sex']=='Male']['total_bill'],
            tips[tips['sex']=='Male']['tip'],
            label='Male', color='blue')

plt.scatter(tips[tips['sex']=='Female']['total_bill'],
            tips[tips['sex']=='Female']['tip'],
            label='Female', color='red')

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects,autopct='%0.0f%%')

plt.show()
```

```python
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects,autopct='%0.1f%%',colors=['blue','green','yellow','pink','brown'])

plt.show()
```

```python
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects,autopct='%0.1f%%',colors=['blue','green','yellow','pink','brown'],explode=[0.1,0,0,0,0.1])

plt.show()
plt.savefig('sample.png')
```

```python
plt.subplot() is used in Matplotlib to create multiple plots in a single figure (subplots).
```

```python
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(df['avg'],df['strike_rate'],color='red')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,6))
print(ax)
```
